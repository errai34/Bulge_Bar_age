```python
import pandas as pd

%matplotlib inline
```

```python
df = pd.read_csv('../data/7dotS39_ALL_nocuts_extendedFMT.txt', skiprows=1)

def process_andrea_data(data_path):
    
    df = pd.read_csv(data_path, skiprows=1)
    
    if data_path == "../data/7dotS39_ALL_FiorenzoFMT.txt":
    
        vals = [df.iloc[:, 0].loc[x].split() for x in df.index]
        kic = [vals[x][0] for x in range(len(vals))]
      # ra = [vals[x][1] for x in range(len(vals))] this is not ra, dec
      # dec = [vals[x][2] for x in range(len(vals))]
        feh = [vals[x][3] for x in range(len(vals))]
        afe = [vals[x][4] for x in range(len(vals))]
        age = [vals[x][5] for x in range(len(vals))]
        age_68l = [vals[x][6] for x in range(len(vals))]
        age_68u = [vals[x][7] for x in range(len(vals))]
        mass = [vals[x][8] for x in range(len(vals))]
        mass_68l = [vals[x][9] for x in range(len(vals))]
        mass_68u = [vals[x][10] for x in range(len(vals))]
        evstate = [vals[x][11] for x in range(len(vals))]
        
        new_df = pd.DataFrame()

        new_df['KIC'] = kic
       # new_df['RA'] = ra
       # new_df['DEC'] = dec
        new_df['feh'] = feh
        new_df['afe'] = afe
        new_df['age'] = age
        new_df['age_68L'] = age_68l
        new_df['age_68U'] = age_68u
        new_df['mass'] = mass
        new_df['mass_68L'] = mass_68l
        new_df['mass_68U'] = mass_68u
        new_df['evstate'] = evstate
        
    elif data_path == "../data/7dotS39_ALL_nocuts_extendedFMT.txt":
        
        vals = [df.iloc[:, 0].loc[x].split() for x in df.index]
        
        apogee_id = [vals[x][0] for x in range(len(vals))]
        kic = [vals[x][1] for x in range(len(vals))]
        ra = [vals[x][2] for x in range(len(vals))]
        dec = [vals[x][3] for x in range(len(vals))]
        feh = [vals[x][4] for x in range(len(vals))]
        afe = [vals[x][5] for x in range(len(vals))]
        age = [vals[x][6] for x in range(len(vals))]
        age_68l = [vals[x][7] for x in range(len(vals))]
        age_68u = [vals[x][8] for x in range(len(vals))]
        mass = [vals[x][9] for x in range(len(vals))]
        mass_68l = [vals[x][10] for x in range(len(vals))]
        mass_68u = [vals[x][11] for x in range(len(vals))]
        dist = [vals[x][12] for x in range(len(vals))]
        dist_68L = [vals[x][13] for x in range(len(vals))]
        dist_68U = [vals[x][14] for x in range(len(vals))]
        Av = [vals[x][15] for x in range(len(vals))]
        Av_68L = [vals[x][16] for x in range(len(vals))]
        Av_68U = [vals[x][17] for x in range(len(vals))]
        evstate = [vals[x][18] for x in range(len(vals))]
        
        new_df = pd.DataFrame()
        new_df['APOGEE_ID'] = apogee_id
        new_df['KIC'] = kic
        new_df['RA'] = ra
        new_df['DEC'] = dec
        new_df['feh'] = feh
        new_df['afe'] = afe
        new_df['age'] = age
        new_df['age_68L'] = age_68l
        new_df['age_68U'] = age_68u
        new_df['mass'] = mass
        new_df['mass_68L'] = mass_68l
        new_df['mass_68U'] = mass_68u      
        new_df['dist'] = dist
        new_df['dist_68L'] = dist_68L
        new_df['dist_68U'] = dist_68U
        new_df['Av'] = Av
        new_df['Av_68L'] = Av_68L
        new_df['Av_68U'] = Av_68U
        new_df['evstate'] = evstate
        
    else:
        pass
    
    return new_df
        
```

```python
nocut = "../data/7dotS39_ALL_nocuts_extendedFMT.txt"
cut = "../data/7dotS39_ALL_FiorenzoFMT.txt"
```

```python
cut_andrea = process_andrea_data(cut)
nocut_andrea = process_andrea_data(nocut)

merged = pd.merge(cut_andrea, nocut_andrea, on="KIC")

df_merged = merged.drop(['feh_y', 'afe_y', 'age_y', 'age_68L_y',
       'age_68U_y', 'mass_y', 'mass_68L_y', 'mass_68U_y', 'evstate_y'], axis=1)

clean_df = pd.DataFrame()

clean_df["APOGEE_ID"] = df_merged["APOGEE_ID"]
clean_df["KIC"] = df_merged["KIC"]
clean_df["RA"] = df_merged["RA"]
clean_df["DEC"] = df_merged["DEC"]
clean_df["feh"] = df_merged["feh_x"]
clean_df["afe"] = df_merged["afe_x"]
clean_df["age"] = df_merged["age_x"]
clean_df["age_68L"] = df_merged["age_68L_x"]
clean_df["age_68U"] = df_merged["age_68U_x"]
clean_df["mass"] = df_merged["mass_x"]
clean_df["mass_68L"] = df_merged["mass_68L_x"]
clean_df["mass_68U"] = df_merged["mass_68U_x"]
clean_df["dist"] = df_merged["dist"]
clean_df["dist_68L"] = df_merged["dist_68L"]
clean_df["dist_68U"] = df_merged["dist_68U"]
clean_df["Av"] = df_merged["Av"]
clean_df["Av_68L"] = df_merged["Av_68L"]
clean_df["Av_68U"] = df_merged["Av_68U"]
clean_df["evstate"] = df_merged["evstate_x"]
```

```python
cols_of_int = ['RA', 'DEC', 'feh', 'afe', 'age', 'age_68L',
               'age_68U', 'mass', 'mass_68L', 'mass_68U', 'dist', 'dist_68L',
               'dist_68U', 'Av', 'Av_68L', 'Av_68U', 'evstate']
```

```python
clean_df[cols_of_int] = clean_df[cols_of_int].apply(pd.to_numeric)
```
